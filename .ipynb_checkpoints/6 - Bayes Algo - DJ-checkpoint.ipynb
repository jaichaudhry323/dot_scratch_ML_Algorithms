```python
import numpy as np
import pandas as pd
eps = np.finfo(float).eps
```

```python
eps
```

```python
def findRes(df, query):
    Class = df.keys()[-1]
    attributes = df.columns
    target_variables = df[Class].unique()
    
    probabilities = []
    
    for target_variable in target_variables:
        
        den = len(df[Class][df[Class]==target_variable])
        total = df.shape[0]
        
        prob = den/total
        #print(den,"/",total)
        
        for i in range(len(query)):
            attr = attributes[i]
            num = len(df[attr][df[attr]==query[i]][df[Class]==target_variable])
            frac = num/den
            #print(num,"/",den)
            prob *= frac
            
        probabilities.append(prob)
        #print(prob)
        
    return target_variables[np.argmax(probabilities)]


def NaiveBayes(train_df, test_df, printQueryResults=False):
    test_df = test_df.reset_index(drop=True)
    correct = 0
    
    for i in range(test_df.shape[0]):
        query = test_df.iloc[i,:test_df.shape[1]-1]
        res = findRes(train_df, query)
        actual = test_df.iloc[i,:][-1]
        
        if printQueryResults==True:
            print("predicted : ",res, " actual : ", actual)
        
        if res==actual:
            correct += 1
            
    acc = correct*100/test_df.shape[0]
    print("accuracy : ", acc)
        
            
```

```python
#encoding the age column by making 4 categories => 0:Children, 1:Youth, 2:Adults, 3:Elderly

ageCol = df['age'].copy()
ageCol[df['age']<=14] = 0 
ageCol[(df['age']>14) & (df['age']<=24)] = 1
ageCol[(df['age']>24) & (df['age']<=64)] = 2
ageCol[(df['age']>64)] = 3

df['age'] = ageCol
```

```python
col = ["fnlwgt", "educational_num", "capital_gain", "capital_loss", "hours_per_week"]

#encoding these columns in following way => 0 : value<mean, 1 : value>=mean

for c in col:
    
    temp = df[c].copy()
    avg = np.mean(df[c])
    
    temp[df[c]<avg] = 0
    temp[df[c]>=avg] = 1
    
    df[c] = temp
```
